```python
from kuibit import simdir as sd

name="30"

with sd.SimDir("/home/yuntian/vscode_codes/QNM_compare/Kerr_Newman/"+name,pickle_file="/home/yuntian/vscode_codes/QNM_compare/Kerr_Newman/"+name+".pickle") as sim:
    print(sim)
```

```python
psi4=sim.multipoles["psi4"][110.69]
print(psi4.available_lm)
```

```python
for lm in psi4.available_lm:
    l=lm[0]
    m=lm[1]
    fo=open(f"NR_data_"+name+f"/Weyl4_mode_0_{l}{m}.dat","w")
    fo.write("#       time         integral Re         integral Im\n")
    fo.write("#       r =           110.690000          110.690000\n")

    psi4_lm=psi4[l,m]


    for i in range(len(psi4_lm.t)):
        t=psi4_lm.t[i]
        psi4_r=psi4_lm.values[i].real
        psi4_i=psi4_lm.values[i].imag
        fo.write(f"{t:0>12.7f}   {psi4_r:+.10e}   {psi4_i:+.10e}\n")

    fo.close()
```
